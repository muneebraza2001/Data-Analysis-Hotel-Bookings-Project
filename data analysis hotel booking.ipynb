DATA ANALYSIS AND VISUALIZATIONS

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc

plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not cancled' , 'Cancled'], df['is_canceled'].value_counts() , edgecolor = 'k' , width = 0.7)
plt.show()
```

The graph above shows the percentage of reservations that are canceled and those that are not. It is shown that there are still a significant number of reservations that have not been canceled. There are still few of clients who canceled their reservation, which has a significant impact on the hotels' earnings

```python
plt.figure(figsize = (8,4))
axl = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

the graph above shows that in comparasion to resort hotels,city hotels have more bookings .its oissible that resort hotels are more expensive than for ones in cities.

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(15, 6))  
plt.title('Average Normal Day Rate in City and Resort Hotels', fontsize=25)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=15)
plt.show()
```

the line graph above shows that on certain days the average daily rate for a city hotel is less than of resort hotel and on ther days it is more low. it goes without saying that weekends and holidays may see a rise in resort hotel rates.

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (15,6))
axl = sns.countplot(x = 'month', hue='is_canceled', data= df, palette = 'bright')
legend_labels,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not_canceled','canceled'])
plt.show()


```

the graph above to analyze the months with the highest and lowest reservation levels according to reservation status. As can be seen both the number of confirmed reservations and the number of canceled reservations are largest in the month of August. whereas January is the month with the most canceled reservations

```python
plt.figure(figsize=(15, 6))
plt.title('ADR per Month', fontsize=20)

sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

plt.xlabel('Month')
plt.ylabel('ADR (Average Daily Rate)')
plt.show()
```

This graph above shows that cancellations are most common when prices are greatest and are least common when they are lowest. Therefore, the cost of the accommodation is solely responsible for the cancellation.

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8, 8))
plt.title('Top 10 Countries with Reservations Canceled')
plt.pie(top_10_country, autopct='%.2f%%', labels=top_10_country.index)

plt.show()
```

the chart above shows the area from where guests are visiting the hotels and making reservations. Is it coming from Direct or Groups, Online or Offline Travel Agents? Around 46% of the clients come from online travel agencies, whereas 27% come from groups. Only 4% of clients book hotels directly by visiting them and making reservations.

```python
# average ADR for canceled reservations
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)  
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# average ADR for non-canceled reservations
not_cancelled_data = df[df["is_canceled"] == 0]  
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(15, 6)) 
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.legend()
plt.show()
```

the graph above shows that the reservations are canceled when the average daily rate is higher than when it is not canceled which proves all the above analysis, that the higher price leads to higher cancellation.
